# Kafka to Databricks Streaming (Topper Level)


```
df = spark.readStream.format('kafka')\
.option('kafka.bootstrap.servers','localhost:9092')\
.option('subscribe','smartgear_orders')\
.load()
```

```
from pyspark.sql.functions import from_json, col
schema = 'order_id INT, timestamp STRING, store_id INT, product STRING, quantity INT, price INT, region STRING'
df_parsed = df.selectExpr('CAST(value AS STRING)')\
.select(from_json(col('value'), schema).alias('data')).select('data.*')
```

```
df_parsed.writeStream.format('delta')\
.option('checkpointLocation','/mnt/checkpoints')\
.outputMode('append')\
.start('/mnt/bronze_stream')
```
