# SmartGear Full Pipeline


```
# Bronze Layer
df = spark.read.format('csv').option('header','true').load('/FileStore/smartgear_sales.csv')
df.write.format('delta').mode('append').save('/mnt/bronze')
```

```
# Silver Layer
df = spark.read.format('delta').load('/mnt/bronze')
df = df.dropDuplicates()
df = df.withColumn('revenue', df.quantity * df.price)
df.write.format('delta').mode('overwrite').save('/mnt/silver')
```

```
# Gold Layer
df = spark.read.format('delta').load('/mnt/silver')
df.groupBy('region').sum('revenue').show()
```
