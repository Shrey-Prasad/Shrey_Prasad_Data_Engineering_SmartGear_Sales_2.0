# Advanced SQL - Gold Layer


```
-- Top Products
SELECT product, SUM(revenue) as total_rev,
RANK() OVER (ORDER BY SUM(revenue) DESC) as rank
FROM silver GROUP BY product;
```

```
-- Rolling Avg
SELECT date, AVG(revenue) OVER (ORDER BY date ROWS BETWEEN 6 PRECEDING AND CURRENT ROW)
FROM silver;
```

```
-- MoM Growth
SELECT month, SUM(revenue),
LAG(SUM(revenue)) OVER (ORDER BY month) as prev,
(SUM(revenue)-LAG(SUM(revenue)) OVER (ORDER BY month))/LAG(SUM(revenue)) OVER (ORDER BY month) as growth
FROM silver GROUP BY month;
```
